```python
time_steps = [['1990', '2000'],
              ['2000', '2010'],
              ['2010', '2015'],
              ['2015', '2020']] # 4개의 시간범위를 이중리스트를 사용해 time_steps 변수에 저장한다.

fig, axes = plt.subplots(2, 2) # fig와 axes변수에 2x2 크기의 서브플롯을 저장
fig.set_size_inches(16, 9) # 그래프 전체의 가로세로 인치수를 설정한다.
for i in range(4): # 반복문 사용
    ax = axes[i//2, i%2]
    df = stock.loc[(stock.index > time_steps[i][0]) & (stock.index < time_steps[i][1])] # &연산자를 활용한 조건에 충족할 시 저장되는 행으로, &연산자는 양 옆의 조건이 만족하는 경우 해당 데이터를 저장하는 행
    # 왼쪽 조건 / stock.index값이 time_steps[i][0](time_steps리스트의 i번째 요소의 0번째 요소인 스타트 시각)보다 클때 / 오른쪽 조건 / stock.index값이 time_steps[i][1](time_steps리스트의 i번째 요소의 1번째 요소인 엔드 시각)보다 작을 때
    sns.lineplot(y=df['Close'], x=df.index, ax=ax) # y축은 df['Close']에 관한, x축은 df.index에 관한 그래프를 그린다. (ax=ax 는 그릴 서브플롯을 지정한다?)
    ax.set_title(f'{time_steps[i][0]}~{time_steps[i][1]}') # 해당 그래프의 제목은 시간 범위로 지정
    ax.set_xlabel('time') # xlabel을 time으로, ylabel을 price로 지정
    ax.set_ylabel('price')
plt.tight_layout()
plt.show() # 그래프 출력
```

```python
from sklearn.preprocessing import MinMaxScaler #최대 최소값 스케일링

scaler = MinMaxScaler()

scale_cols = ['Open', 'High', 'Low', 'Close', 'Volume'] # 스케일링을 할 데이터가 속한 열을 설정

scaled = scaler.fit_transform(stock[scale_cols]) # 위에서 설정한 열에 따라 스케일링을 진행하고 나온 데이터를 scaled 변수에 저장함
scaled
```

```python
from sklearn.model_selection import train_test_split # 데이터를 학습세트와 테스트 세트로 나누는 용도?
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Close', 1), df['Close'], test_size=0.2, random_state=0, shuffle=False)
# 앞서 선언한 데이터프레임 df에서 Close열을 제외한 모든 열의 데이터와 df['Close']를 사용해 학습 세트와 테스트 세트로 나눕니다. (train 학습 / test 테스트)
# 해당 테스트 세트의 반영 비율은 20%이며 , 난수를 발생시키는 시드를 0으로 고정한다(?) , 또한 데이터를 섞지 않는다.
```

```python
import tensorflow as tf # 딥러닝에 쓰이는 라이브러리
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle): ##시계열 데이터 (일정시간동안 순차적으로 수집한 데이터)를 일정한 구간으로 나누어 각 구간을 하나의 샘플로 저장시키는 함수
    series = tf.expand_dims(series, axis=-1) ##Tensorflow의 RNN모델(시간에 따라 변화하는 데이터를 처리하기 위한 인공 신경망)에 입력하기 위해 차원을 하나 추가함
    ds = tf.data.Dataset.from_tensor_slices(series) ##series의 데이터로부터 TensorFlow 데이터셋(ds)을 생성합니다
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    if shuffle:
        ds = ds.shuffle(1000)
    ds = ds.map(lambda w: (w[:-1], w[-1]))
    return ds.batch(batch_size).prefetch(1)
```

```python
WINDOW_SIZE=20
BATCH_SIZE=32
```

```python

train_data = windowed_dataset(y_train, WINDOW_SIZE, BATCH_SIZE, True)
test_data = windowed_dataset(y_test, WINDOW_SIZE, BATCH_SIZE, False)
```

```python
##아래의 코드로 데이터셋의 구성을 확인해 볼 수 있습니다.
for data in train_data.take(1):
    print(f'데이터셋(X) 구성(batch_size, window_size, feature갯수): {data[0].shape}')
    print(f'데이터셋(Y) 구성(batch_size, window_size, feature갯수): {data[1].shape}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, Lambda
from tensorflow.keras.losses import Huber
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint


model = Sequential([
    # 1차원 feature map 생성
    Conv1D(filters=32, kernel_size=5,
           padding="causal",
           activation="relu",
           input_shape=[WINDOW_SIZE, 1]),
    # LSTM (16개의 유닛을 가지며, 활성화 함수로는 쌍곡선 탄젠트(tanh) 함수를 사용합니다.)

    LSTM(16, activation='tanh'),
    Dense(16, activation="relu"),
    Dense(1),
])
```

```python
##Sequence 학습에 비교적 좋은 퍼포먼스를 내는 Huber()를 사용합니다.
loss = Huber()
optimizer = Adam(0.0005)
model.compile(loss=Huber(), optimizer=optimizer, metrics=['mse'])
```

```python
##earlystopping은 10번 epoch통안 val_loss 개선이 없다면 학습을 멈춥니다.
earlystopping = EarlyStopping(monitor='val_loss', patience=10)
##val_loss 기준 체크포인터도 생성합니다.
filename = os.path.join('tmp', 'ckeckpointer.ckpt')
checkpoint = ModelCheckpoint(filename,
                             save_weights_only=True,
                             save_best_only=True,
                             monitor='val_loss',
                             verbose=1)
```

```python
history = model.fit(train_data,
                    validation_data=(test_data),
                    epochs=50,
                    callbacks=[checkpoint, earlystopping])
```
